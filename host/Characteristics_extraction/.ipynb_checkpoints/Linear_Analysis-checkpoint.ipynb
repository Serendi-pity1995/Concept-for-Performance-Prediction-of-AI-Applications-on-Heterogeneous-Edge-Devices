```python
import pandas
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
df = pandas.read_csv("training_data+valid_data.csv")
x1 = df["FLOPs_G"]
x2 = df["Total_params_M"]
# x3 = df["image_shape_x"]
x4 = df["Memory_M"]
x5 = df["total_num_of_neurons_M"]
y = df["Execution_time"]
result = sm.ols(formula="Execution_time ~ FLOPs_G + Total_params_M + Memory_M ", data=df).fit()
print (result.summary())
print (result.rsquared, result.rsquared_adj)
plt.xlabel("Total number of FLOPs(G)")
plt.ylabel("Execution time(ms)")
plt.title("Total number of FLOPs vs. Execution time ")
plt.scatter(x1, y)
print(result.params)
```
