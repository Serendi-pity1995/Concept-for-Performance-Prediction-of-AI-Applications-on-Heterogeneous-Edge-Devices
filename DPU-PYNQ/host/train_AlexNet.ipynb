```python
import numpy as np
import os
import keras
from keras import backend as K

import random
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D, Input, BatchNormalization
from tensorflow_model_optimization.quantization.keras import vitis_quantize
from keras.models import Model
import tensorflow as tf



shape_range =range(224,512,32)
for shape_x in shape_range:
    shape_y = shape_x
    X_train=np.random.randint(0,255,size=[60,shape_x,shape_y,1])
    Y_train=np.random.randint(0,10,size=[60,])
    X_test=np.random.randint(0,255,size=[60,shape_x,shape_y,1])
    Y_test=np.random.randint(0,10,size=[60,])
    X_train = X_train.astype('float32')
    X_test = X_test.astype('float32')
    X_train /= 255.0
    X_test /= 255.0
    inputs = Input(shape=(shape_x, shape_y, 1))
    for step in range(1,11):
    
        conv1_kernel_num = random.randrange(64,96,2)
        conv1 = Conv2D(conv1_kernel_num, (11, 11), strides=(4, 4), activation='relu')(inputs)
        pool1 = MaxPooling2D(pool_size=(3, 3), strides=(2, 2))(conv1)
        Norm1 = BatchNormalization()(pool1)

        conv2_kernel_num = random.randrange(128,256,2)
        conv2 = Conv2D(conv2_kernel_num, kernel_size=(5, 5), activation='relu')(Norm1)
        pool2 = MaxPooling2D(pool_size=(3, 3), strides=(2, 2))(conv2)
        Norm2 = BatchNormalization()(pool2)

        conv3_kernel_num = random.randrange(conv2_kernel_num,384,2)
        conv3 = Conv2D(384, kernel_size=(3, 3), activation='relu')(Norm2)
        conv4_kernel_num = random.randrange(conv3_kernel_num,384,2)
        conv4 = Conv2D(384, kernel_size=(3, 3), activation='relu')(conv3)
        conv4_kernel_num = random.randrange(128,256,2)
        conv5 = Conv2D(conv4_kernel_num, kernel_size=(3, 3), activation='relu')(conv4)
        pool3 = MaxPooling2D(pool_size=(3, 3), strides=(2, 2))(conv5)
        Norm3 = BatchNormalization()(pool3)

        flat = Flatten()(Norm3)
        Dense1_neuron_num = random.randrange(1072,3072,32)
        fc1 = Dense(Dense1_neuron_num, activation='relu')(flat)
        drop1 = Dropout(0.5)(fc1)
        Dense2_neuron_num = random.randrange(Dense1_neuron_num,3072,32)
        fc2 = Dense(Dense2_neuron_num, activation='relu')(drop1)
        drop2 = Dropout(0.5)(fc2)

        outputs = Dense(10, activation='softmax')(drop2)

        model = Model(inputs=inputs, outputs=outputs)
        model.summary()
        model.save('./AlexNet_h5/AlexNet_{input_x}_{input_y}_{step}.h5'.format(input_x=shape_x,input_y=shape_y,step=step))

    #     model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    #     history = model.fit(X_train, Y_train, epochs=1)
    #     model.get_weights()[0].dtype
        quantizer = vitis_quantize.VitisQuantizer(model)
        quantized_model = quantizer.quantize_model(calib_dataset = X_test[1:10], weight_bit=8, activation_bit=8)
    #     quantized_model.compile(loss='sparse_categorical_crossentropy', metrics=["accuracy"])
        quantized_model.save('quantized.h5')
        !vai_c_tensorflow2 \
        --model ./quantized.h5 \
        --arch ./arch.json \
        --output_dir ./xmodel_AlexNet \
        --net_name deploy
        os.rename("./xmodel_AlexNet/deploy.xmodel","./xmodel_AlexNet/AlexNet_{shape_x}_{shape_y}_{step}.xmodel".format(shape_x=shape_x,shape_y=shape_y,step=step))
```
