```python
import pandas
import numpy as np
#import statsmodels.formula.api as sm

import matplotlib.pyplot as plt # for data visualization
%matplotlib inline
df = pandas.read_csv("pred_AlexNet.csv")
x = df["model_num"]
y_pred = df["2_variables_model"]
y_test = df["Execution time"]

plt.xlabel("index")
plt.ylabel("execution time")
plt.title("prediction result(red points) vs real execution time(green points)")
plt.scatter(x, y_pred,c="r")
plt.scatter(x, y_test,c="g")


# index=np.argsort(y_test)

# plt.figure(figsize=(12,5))
# plt.plot(np.arange(len(y_test)),y_test[index],"r",label="original y")
# plt.scatter(np.arange(len(y_pred)),y_pred[index],s=3,c="b",label="prediction")
# plt.legend(loc="upper left")
# plt.grid()
# plt.xlabel("index")
# plt.ylabel("y")
# plt.show()
```

```python
import pandas
import statsmodels.formula.api as sm
import matplotlib.pyplot as plt
df = pandas.read_csv("pred_AlexNet.csv")
x = df["model_num"]
y = df["2_variables_error(%)"]

plt.xlabel("index")
plt.ylabel("error(%)")
plt.title("2_variables_model prediction error ")
plt.bar(x, y)
```
